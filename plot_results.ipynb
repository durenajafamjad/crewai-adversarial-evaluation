```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the infections log JSON file
file_path = "infection_rate_logs.json"  # Update with your file path
with open(file_path, "r") as f:
    infections_data = json.load(f)

# Process the data into a structured format
data = []
for entry in infections_data:
    agent_name = entry["file_name"].split("/")[1]  # Extract agent name (e.g., accountant)
    testing_model = entry["testing_model"]
    response_model = entry["response_model"]
    total_queries = entry["total_queries"]
    harmful_count = entry["harmful_count"]
    infected_rate = float(entry["infected_rate"].strip("%"))  # Convert percentage to float

    data.append({
        "Agent": agent_name,
        "Testing Model": testing_model,
        "Response Model": response_model,
        "Total Queries": total_queries,
        "Harmful Count": harmful_count,
        "Infected Rate (%)": infected_rate
    })

# Convert the data into a DataFrame
df = pd.DataFrame(data)

# Plot 1: Infection Rates for Testing Models across Agents
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x="Testing Model", y="Infected Rate (%)", hue="Agent")
plt.title("Harmful Rates (%) by Testing Models Across Agents")
plt.ylabel("Harmful Rate (%)")
plt.xlabel("Testing Model")
plt.xticks(rotation=45)
plt.legend(title="Agent")
plt.tight_layout()
plt.show()

# Plot 2: Harmful Counts for Response Models across Agents
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x="Response Model", y="Harmful Count", hue="Agent", palette="viridis")
plt.title("Harmful Query Counts by Response Models Across Agents")
plt.ylabel("Harmful Query Count")
plt.xlabel("Response Model")
plt.xticks(rotation=45)
plt.legend(title="Agent")
plt.tight_layout()
plt.show()

# Plot 3: Total Queries per Agent and Model
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x="Agent", y="Total Queries", hue="Testing Model", palette="coolwarm")
# plt.title("Total Queries Answered by Agents and Testing Models")
plt.ylabel("Total Queries")
plt.xlabel("Agent")
plt.legend()
plt.tight_layout()
plt.show()
```
